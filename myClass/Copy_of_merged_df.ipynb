```python
def or_func(x):
  comp = ['comp1','comp2','comp3','comp4']
  if x['comp'] in comp:
    return x['comp']
  elif x['failure'] in comp:
    return x['failure']
  else:
    return None
```

```python
merged_df['label'] = merged_df.apply(or_func,axis=1)
```

```python
merged_df.drop(['comp','failure'],axis=1,inplace=True)
```

```python
merged_df
```

```python
merge_tel = pd.merge(df_tel, merged_df, on=['datetime', 'machineID'], how='left', suffixes=('_df1', '_df2'))
```

```python
merge_tel.head()
```

```python
def add_label(x):
  comp = ['comp1','comp2','comp3','comp4']
  if x['label'] not in comp:
    return "no_comp"
  return x['label']
```

```python
merge_tel['label'] = merge_tel.apply(add_label,axis=1)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error


under_sampled_df['day_of_week'] = under_sampled_df['time'].dt.dayofweek
under_sampled_df['month'] = under_sampled_df['time'].dt.month
under_sampled_df['year'] = under_sampled_df['time'].dt.year


under_sampled_df = under_sampled_df.drop(columns=['time'])


X = under_sampled_df.drop(['machineID'], axis=1)

y1 = under_sampled_df['machineID']
y2 = under_sampled_df['month']


X_train, X_test, y1_train, y1_test, y2_train, y2_test = train_test_split(X, y1, y2, test_size=0.2, random_state=42)


rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)


rf_regressor.fit(X_train, list(zip(y1_train, y2_train)))


y_pred = rf_regressor.predict(X_test)


y1_pred, y2_pred = zip(*y_pred)


mse_y1 = mean_squared_error(y1_test, y1_pred)
mse_y2 = mean_squared_error(y2_test, y2_pred)


```

```python
r2_y1 = r2_score(y1_test, y1_pred)

# Calculate R-squared for y2
r2_y2 = r2_score(y2_test, y2_pred)

print("R-squared for Target Variable 1:", r2_y1)
print("R-squared for Target Variable 2:", r2_y2)
```
