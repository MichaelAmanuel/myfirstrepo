```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('white')
```

```python
columns_name = ['user_id', 'item_id', 'rating', 'timestamp']
```

```python
data = pd.read_csv('/content/u.data', sep='\t', names = columns_name)
```

```python
data.head()
```

```python
movie_titles = pd.read_csv('/content/Movie_Id_Titles')
```

```python
movie_titles.head()
```

```python
data = pd.merge(data, movie_titles, on='item_id')
```

```python
data.head()
```

```python
data.groupby('title')['rating'].mean()
```

```python
data.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
data.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
ratings = pd.DataFrame(data.groupby('title')['rating'].mean())
```

```python
ratings.head()
```

```python
ratings['num of ratings'] = pd.DataFrame(data.groupby('title')['rating'].count())
```

```python
ratings.head()
```

```python
ratings['num of ratings'].hist(bins=70)
```

```python
sns.histplot(ratings, kde=False,bins=70)
```

```python
plt.figure(figsize=(11, 6))
ratings['rating'].hist(bins=70)
```

```python
sns.jointplot(x='rating', y='num of ratings', data=ratings, alpha=0.5)
```

```python
data.head()
```

```python
moviemat = data.pivot_table(index='user_id', columns='title',values='rating')
```

```python
moviemat.head()
```

```python
ratings.sort_values('num of ratings', ascending=False).head()
```

```python
starwars_user_ratings = moviemat['Star Wars (1977)']
liarlair_user_ratings = moviemat['Liar Liar (1997)']
```

```python
starwars_user_ratings.head()
```

```python
liarlair_user_ratings.head()
```

```python
moviemat.corrwith(starwars_user_ratings)
```

```python
similar_to_starwars = moviemat.corrwith(starwars_user_ratings)
```

```python
similar_to_liarliar = moviemat.corrwith(liarlair_user_ratings)
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars, columns=['Correlation'])
corr_starwars.dropna(inplace=True)
```

```python
corr_starwars.head()
```

```python
corr_starwars.sort_values('Correlation', ascending=False).head(10)
```

```python
corr_starwars = corr_starwars.join(ratings['num of ratings'], rsuffix='_num_of_ratings')
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['num of ratings'] > 100].sort_values(
                                                                'Correlation',
                                                                ascending=False
                                                                ).head()
```

```python
corr_liarliar = pd.DataFrame(similar_to_liarliar, columns=['Correlation'])
corr_liarliar.dropna(inplace=True)
```

```python
corr_liarliar.head()
```

```python
corr_liarliar = corr_liarliar.join(ratings['num of ratings'], rsuffix='_num_of_ratings')
corr_liarliar.head()
```

```python
corr_liarliar[corr_liarliar['num of ratings'] > 100].sort_values(
                                                                'Correlation',
                                                                ascending=False
                                                                ).head()
```
