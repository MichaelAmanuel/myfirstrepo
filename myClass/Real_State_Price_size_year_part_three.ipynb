```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('/content/real_estate_price_size_year (1).csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
X = data[['size', 'year']]
y = data['price']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
X_scaled = scaler.transform(X)
```

```python
X_scaled
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_scaled, y)
```

```python
reg.score(X_scaled, y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
pred_value = np.round(reg.predict(X_scaled))
```

```python
data['Predict_price'] = pred_value
```

```python
data.head()
```

```python
def adj_r2(x,y):
    r2 = reg.score(x,y)
    n = x.shape[0]
    p = x.shape[1]
    adjusted_r2 = 1-(1-r2)*(n-1)/(n-p-1)
    return adjusted_r2
```

```python
adj_r2(X_scaled,y)
```

```python
new_data = [[750,2009]]
new_data_scaled = scaler.transform(new_data)
```

```python
reg.predict(new_data_scaled)
```
