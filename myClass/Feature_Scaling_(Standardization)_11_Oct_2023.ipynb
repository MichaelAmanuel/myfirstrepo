```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('/content/1.02.+Multiple+linear+regression (1).csv')
data.head()
```

```python
data.describe()
```

```python
X = data[['SAT', 'Rand 1,2,3']]
y = data['GPA']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
x_scaled = scaler.transform(X)
```

```python
x_scaled
```

```python
reg = LinearRegression()
```

```python
reg.fit(x_scaled, y)
```

```python
reg.coef_
```

```python
reg.score(x_scaled, y)
```

```python
reg.intercept_
```

```python
scaled_gpa = np.round(reg.predict(x_scaled), 2)
```

```python
data['Predicted_GPA'] = scaled_gpa
```

```python
data.head()
```

```python
reg_summary = pd.DataFrame([['Intercept'], ['SAT'], ['Rand 1,2,3']], columns=['Features'])
reg_summary['Weight'] = reg.intercept_, reg.coef_[0], reg.coef_[1]
```

```python
reg_summary
```

```python
new_data = pd.DataFrame(data=[[1700, 2], [1800,1]], columns=['SAT', 'Rand 1,2,3'])
```

```python
new_data
```

```python
reg.predict(new_data)
```

```python
new_data_scaled = scaler.transform(new_data)
new_data_scaled
```

```python
reg.predict(new_data_scaled)
```
