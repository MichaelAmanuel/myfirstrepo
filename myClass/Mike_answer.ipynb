```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_ylabel('Y')
ax.set_xlabel('X')
ax.set_title('Title')
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])

ax1.plot(x, y, color='r')
ax2.plot(x, y, color='r')
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax1.set_xlabel('X')
ax1.set_ylabel('Z')
ax2 = fig.add_axes([0.2,0.5,.4,.4])
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
ax2.set_title('Zoom')

ax1.plot(x,z)
ax2.plot(x, y)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)

ax[0].plot(x,y,color='b',lw=3,ls='--')
ax[1].plot(x,z, color='r',lw=3)
```

```python
fig, ax = plt.subplots(1,2)

ax[0].plot(x,y,color='b',lw=3,ls='--')
ax[1].plot(x,z, color='r',lw=3)
```

```python
fig, ax = plt.subplots(1,2, figsize=(12,2))

ax[0].plot(x,y,color='b',lw=3,ls='--')
ax[1].plot(x,z, color='r',lw=3)
```
