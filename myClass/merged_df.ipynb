```python
common_columns = ['datetime', 'machineID']

non_common_cols_df_one = [col for col in df_one.columns if col not in common_columns]
non_common_cols_df_two = [col for col in df_two.columns if col not in common_columns]
```

```python
merged_df = pd.merge(df_one, df_two, on=common_columns, how='outer')

# Concatenate non-common columns
df = pd.concat([merged_df[common_columns], merged_df[non_common_cols_df_one], merged_df[non_common_cols_df_two]], axis=1)
```
