```python
import matplotlib.pyplot as plt 
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, 'r-')
```

```python
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('X vs Y grouph')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r-')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b-')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('X vs. Y')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes_one = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes.plot(x, y)
axes_one.plot(x, y)
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes_one = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes.plot(x, y)
axes.set_title('Larger Plot')

axes_one.plot(y, x**2)
axes_one.set_title('Smaller Plot')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for pic in axes:
    pic.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[1].plot(y, x)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y)
axes[0].set_xlabel('X Label')
axes[0].set_ylabel('Y Label')
axes[0].set_title('First Plot')

axes[1].plot(y, x)
axes[1].set_xlabel('X Label')
axes[1].set_ylabel('Y Label')
axes[1].set_title('Second Plot')

plt.tight_layout()
```

```python
fig = plt.figure(figsize=(3,2), dpi=100)

ax = fig.add_axes([0,0,1,1])
ax.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8,2))

axes[0].plot(x, y)
axes[1].plot(y, x)

plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label='X Cubed')

ax.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label='X Cubed')

ax.legend(loc = 0)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color='orange')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color='orange',lw =3, alpha=0.5,linestyle="--")
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color='orange',lw =3, alpha=0.5,linestyle="--",marker='*')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color='orange',lw =3,markersize=10,alpha=0.5,linestyle="--",marker='*')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color='purple',lw =2,ls='--')

ax.set_xlim([0, 1])
ax.set_ylim([0, 2])
```
