```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.show()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, y)
ax.set_ylabel('y')
ax.set_xlabel('x')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])

ax.plot(x, y, color='r')
ax2.plot(x, y, color= 'r')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])

# ax2.xlim(2, 4)

ax.plot(x, z)
ax.set_xlabel('X')
ax.set_ylabel('Z')

ax2.plot(x, y)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax2.set_title('Zoom')
ax2.set_xlim([20,22])
ax2.set_ylim([30, 50])
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y, ls='--', color='b',lw=3)
axes[1].plot(x, z,  color='r',lw=3)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 2))

axes[0].plot(x, y, ls='--', color='b',lw=3)
axes[1].plot(x, z,  color='r',lw=3)
```
