```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('/content/1.01.+Simple+linear+regression (1).csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
sns.pairplot(data= data, hue='GPA')
```

```python
sns.pairplot(data= data, hue='SAT')
```

```python
X = data['SAT']
y = data['GPA']
```

```python
X.shape
```

```python
X_matrix = X.values.reshape(-1, 1)
```

```python
X_matrix.shape
```

```python
y.shape
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_matrix, y)
```

```python
reg.score(X_matrix, y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict(X_matrix)
```

```python
data['Predicted_GPA'] = np.round(reg.predict(X_matrix), 2 )
```

```python
data
```

```python
sns.pairplot(data= data, hue='GPA')
```

```python
sns.pairplot(data= data, hue='SAT')
```

```python
sns.histplot(data, kde=False, color='blue', bins=5)
```

```python
plt.scatter(X, y)
yhat = reg.coef_ * X_matrix + reg.intercept_
fig = plt.plot(X, yhat)
plt.xlabel('SAT')
plt.ylabel('GPA')
```
