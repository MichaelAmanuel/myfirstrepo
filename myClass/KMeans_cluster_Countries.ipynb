```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/Countries_exercise.csv')
```

```python
data.head()
```

```python
data.describe(include='all')
```

```python
X = data.iloc[:,1:3]
```

```python
X.head()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(4)
```

```python
kmeans.fit(X)
```

```python
kmeans.fit_predict(X)
```

```python
predicted_value = kmeans.fit_predict(X)

predicted = data.copy()
predicted['predicted_cluster'] = predicted_value
predicted
```

```python
plt.scatter(data['Longitude'], data['Latitude'], c= predicted['predicted_cluster'], cmap='rainbow')
plt.xlim(-180, 180)
plt.ylim(-180, 180)
plt.show()
```

```python
wcss = []

for i in range(1, 15):
  kmeans = KMeans(i)
  kmeans.fit(X)
  wcss_itre = kmeans.inertia_
  wcss.append(wcss_itre)
```

```python
wcss
```

```python
number = range(1, 15)
plt.plot(number, wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of cluster')
plt.ylabel('Within-cluster Sum of Squares')
```
