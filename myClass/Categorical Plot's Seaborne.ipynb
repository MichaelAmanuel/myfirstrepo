```python
import seaborn as sns
%matplotlib inline
import numpy as np
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
sns.barplot(x='sex', y='tip', data=tips)
```

```python
sns.barplot(x='sex', y='size',data=tips)
```

```python
sns.barplot(x='sex', y='total_bill', data=tips,estimator=np.std)
```

```python
sns.barplot(x='sex', y='total_bill', data=tips,estimator=np.mean)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips,hue='smoker')
```

```python
sns.boxplot(x='day', y='total_bill', data=tips,hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex',split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex', jitter=True)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips,hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
```

```python
sns.pointplot(x='sex',y='total_bill',data=tips, hue='sex')
```
