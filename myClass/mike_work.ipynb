```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x, y, 'r-')
```

```python
plt.xlabel('X Label')
plt.ylabel('Y-Label')
plt.title('Title')
plt.plot(x, y)
```

```python
plt.subplot(1,2,1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b', marker= '*' )
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.3, 0.8])

axes.plot(x, y)
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y, 'o')
axes1.set_title('Big Plot')

axes2.plot(y, x , 'o')
axes2.set_title('Smaller Plot')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y)
axes[0].set_title('First Plot')

axes[1].plot(y, x)
axes[1].set_title('Second Plot')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y)
axes[0].set_title('First Plot')

axes[1].plot(y, x)
axes[1].set_title('Second Plot')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, 'g')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

# fig    
# plt.tight_layout()
```

```python
fig = plt.figure(figsize=(3,2), dpi= 100)

axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y)
```

```python
fig, axes = plt.subplots(figsize=(8,2), dpi= 100)

axes.plot(x, y)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(y,x, label='Mike',ls= '-.',lw=4)
axes.plot(x,y, label= 'Raesu',ls=':', lw=4)

axes.legend(loc=10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='b', marker='+',lw=5,ls='-.')
ax.plot(x, 2*x, color='y',marker='*',lw=5, ls='--')
```

```python
fig = plt.figure()

a = fig.add_axes([0,0,1,1])
a.plot(x, y,label='Mike',ls= '-.',lw=4,marker='*')
a.plot(x, x/2,label= 'Raesu',ls=':', lw=4, marker='+')

a.set_xlim([0,1])
a.set_ylim([0,2])

a.legend(loc=0)
```
